```python
def sutun_eksik_veri_miktari(data):
    """
    Sütunlarda toplam eksik değerleri ve oranlarını  döndürür. Grafiğini çizer
    """
    eksik_df = pd.DataFrame({'Nitelik':[],'Toplam Eksik Veri Miktarı':[],'Eksik Veri Oranı':[]})
    eksik_veri_toplam = []
    
    for col in data.columns:
        toplam_eksik_veri = int(data[col].isnull().sum())
        eksik_veri_orani = round((toplam_eksik_veri/data.shape[0])*100,2)
        eksik_df = eksik_df.append(dict(zip(eksik_df.columns,[col,toplam_eksik_veri,eksik_veri_orani])),ignore_index=True)
        eksik_veri_toplam.append(toplam_eksik_veri)
    
    fig, ax = plt.subplots(figsize=(35,5))
    plt.bar(data.columns, eksik_veri_toplam)
    print(f"Eksik veri barındıran sütun sayısı:{str(eksik_df[eksik_df['Eksik Veri Oranı']>0.0].shape[0])}")
    return eksik_df
eksik_data_sutun = sutun_eksik_veri_miktari(data)
eksik_data_sutun

```

```python
def satir_eksik_veri_miktari(data):
    """
    Sütunlarda toplam eksik değerleri ve oranlarını  döndürür. Grafiğini çizer
    """
    eksik_df = pd.DataFrame({'Toplam Eksik Veri Miktarı':[],'Eksik Veri Oranı':[]})
    eksik_veri_toplam = []
    toplam_eksik_veri = data.isnull().sum(axis=1)
    eksik_veri_orani = round((toplam_eksik_veri/len(data.columns))*100,2)
    eksik_df['Toplam Eksik Veri Miktarı'] =  toplam_eksik_veri
    eksik_df['Eksik Veri Oranı'] =  eksik_veri_orani
    '''
    Görselleştirme Sıralama yapılması lazım
    
    '''
    print(f"Eksik veri barındıran satır sayısı:{str(eksik_df[eksik_df['Eksik Veri Oranı']>0.0].shape[0])}")
    print(f"%50 ten fazla eksik veri barındıran sütun sayısı:{str(eksik_df[eksik_df['Eksik Veri Oranı']>50.0].shape[0])}")
    return eksik_df

eksik_data_satir=satir_eksik_veri_miktari(data)
eksik_data_satir
```

- Tam Gözlem Yöntemi (complete case method)
- Mevcut Gözlem Yöntemi (available case method)
- Mevcut Öğe Yöntemi (available item method)
- Birseysel Büyüme Eğirisi Analizi (individual growth curve analysis)
- Çok Örnekli Analiz (multisample analysis) 

şeklinde gösterilmiştir.

## Eksik veri Doldurma Yöntemleri


Bilgisayar ile eksik değerlerin tamamlanması için yaygın olarak iki için teknik kullanılmaktadır

- kutu (Boxplot) Grafiklerine bakılabilir
- saçılım (Scatter) plot grafiğine bakılır
- Z-Skor
- IQR Skoru
- Uzman Görüşü



Outlier'dan kurulmak için 
- Silme
- Ortalama Değer ile doldurma
- Baskılama Yöntemi (üst değer yada alt değer ile doldurma)
gibi işlemler yapılır

#### IQR Skoru

Çeyrekler arası aralık (IQR), orta yayılma veya orta% 50 veya teknik olarak H yayılım olarak da adlandırılır, istatistiksel dağılımın bir ölçüsüdür, 75'inci ve 25'inci yüzdelikler arasındaki veya üst ve alt çeyrekler arasındaki farka eşittir, 

IQR = Q3 - Q1


Başka bir deyişle, IQR, üçüncü çeyrekten çıkarılan ilk çeyrektir; bu çeyrekler, veriler üzerindeki bir kutu grafiği üzerinde açıkça görülebilir.
Standart sapmaya veya varyansa benzer bir dağılım ölçüsüdür, ancak aykırı değerlere karşı çok daha sağlamdır.

```python
Q1 = data_imputed.quantile(0.25)
Q3 = data_imputed.quantile(0.75)
IQR = Q3-Q1
alt_sinir = Q1 - 1.5*IQR
ust_sinir = Q3 + 1.5*IQR
```

```python
print(alt_sinir)
```

```python
print((data_imputed < alt_sinir) | (data_imputed > ust_sinir) )
```

```python
data_outlier_silindi = data_imputed[~((data_imputed < alt_sinir) |(data_imputed > ust_sinir)).any(axis=1)]
```
